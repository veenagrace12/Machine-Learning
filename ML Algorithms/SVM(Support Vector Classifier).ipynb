```python
col_names=['job','marital','education','default','housing','loan','contact','month','poutcome','deposit']
for col in col_names:
    print(df[col].value_counts())
```

```python
# Converting negative values into positive by using  
df['pdays']=abs(df['pdays'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
df["job"] = ord_enc.fit_transform(df[["job"]])
df["marital"] = ord_enc.fit_transform(df[["marital"]])
df["education"] = ord_enc.fit_transform(df[["education"]])
df["default"] = ord_enc.fit_transform(df[["default"]])
df["housing"] = ord_enc.fit_transform(df[["housing"]])
df["loan"] = ord_enc.fit_transform(df[["loan"]])
df["contact"] = ord_enc.fit_transform(df[["contact"]])
df["month"] = ord_enc.fit_transform(df[["month"]])
df["poutcome"] = ord_enc.fit_transform(df[["poutcome"]])
df["deposit"] = ord_enc.fit_transform(df[["deposit"]])
```

```python
Features = df.iloc[:, :-1].values
Labels = df.iloc[:,-1].values
```

```python
Features=abs(Features)
```

```python
Features.shape
```

```python
Labels=abs(Labels)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif, chi2

for i in range(1,len(df.columns)-1):
    sele=SelectKBest(chi2,k=i)
    c=sele.fit_transform(Features,Labels)
print(c)
```

```python
sk = SelectKBest(chi2, k=7)
Features_best = sk.fit_transform(Features, Labels)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Features_best, Labels) 
```

```python
from sklearn.svm import LinearSVC
ls=LinearSVC()
```

```python
ls.fit(X_train,Y_train)
```

```python
Y_pred=ls.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score

print(accuracy_score(Y_pred,Y_test))
print(classification_report(Y_pred,Y_test))
```
